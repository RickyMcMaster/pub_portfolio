```python
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_ISO_3166_country_codes'
df = pd.read_html(url)
# df2.head()
```

```python
len(df)
```

```python
wiki_countries = df[0]
type(wiki_countries)
```

```python
wiki_countries.columns

names = ['Country name',
'Official state name',
'Sovereignty',
'Alpha-2 code',
'Alpha-3 code',
'Numeric code',
'Subdivision code links',
'Internet ccTLD',]

wiki_countries.columns = names
wiki_countries.head()
```

```python
wiki_countries = wiki_countries[~wiki_countries['Country name'].str.contains(' – See')]
```

```python
wiki_countries.columns = wiki_countries.columns.str.lower().str.replace(' ', '_')
wiki_countries.head()
```
